```python
import pandas as pd

# Create a dictionary with more sample data
data = {
    'Name': ['Satender Kumar', 'Arvind', 'Emily', 'John', 'Michael', 'Sophia', 'Isabella', 'Liam', 'Olivia', 'James'],
    'Age': [24, 26, 27, 32, 34, 32, 28, 40, 31, 38],
    'City': ['London', 'Warsaw', 'New York', 'Toronto', 'San Francisco', 'Berlin', 'Sydney', 'Paris', 'Tokyo', 'Moscow'],
    'Occupation': ['Data Analyst', 'Engineer', 'HR Specialist', 'Project Manager', 'Software Developer', 
                   'Data Scientist', 'Product Manager', 'Consultant', 'Marketing Manager', 'Finance Analyst'],
    'Salary': [95000, 80000, 60000, 95000, 105000, 98000, 90000, 85000, 92000, 88000],
    'Experience (Years)': [5, 7, 4, 20, 10, 8, 6, 15, 9, 12],
    'Degree': ['B.Sc. Computer Science', 'M.Eng. Mechanical', 'B.A. Human Resources', 'MBA', 
               'B.Sc. Computer Engineering', 'Ph.D. Data Science', 'MBA', 'B.Sc. Economics', 'B.A. Marketing', 'M.Sc. Finance'],
    'Company': ['ABC Corp', 'XYZ Ltd', 'Global HR', 'Big Projects Inc.', 'Tech Giants', 'Data Minds', 
                'Innovative Solutions', 'Consulting Group', 'Market Leaders', 'Finance Pros'],
    'Country': ['Canada', 'Poland', 'USA', 'Canada', 'USA', 'Germany', 'Australia', 'France', 'Japan', 'Russia']
}
# Create DataFrame
df = pd.DataFrame(data)
```

```python
# Count unique values in the 'Occupation' column
unique_occupations = df['Occupation'].value_counts()
print(unique_occupations)
```

```python
# Filter data for individuals with a salary greater than $90,000
high_salary_df = df[df['Salary'] > 90000]
print(high_salary_df)
```

```python
# Group by 'Country' and calculate the average salary in each country
avg_salary_by_country = df.groupby('Country')['Salary'].mean()
print(avg_salary_by_country)
```

```python
# Group by 'Country' and calculate the average salary in each country
avg_salary_by_country = df.groupby('Country')['Salary'].mean()
print(avg_salary_by_country)
```

```python
# Add a new column 'Salary in Thousands' (Salary/1000)
df['Salary in Thousands'] = df['Salary'] / 1000
print(df)
```

```python
# Add a new column 'Salary in Thousands' (Salary/1000)
df['Salary in Thousands'] = df['Salary'] / 1000
print(df)
```

```python
# Create a pivot table showing average salary by 'City' and 'Occupation'
pivot_table = pd.pivot_table(df, values='Salary', index='City', columns='Occupation', aggfunc='mean')
print(pivot_table)
```

```python
import matplotlib.pyplot as plt

# Plot the distribution of salaries
df['Salary'].hist(bins=10, edgecolor='black')
plt.title('Salary Distribution')
plt.xlabel('Salary')
plt.ylabel('Frequency')
plt.show()
```

```python
# Calculate mean and standard deviation using NumPy
mean_salary = np.mean(df['Salary'])
std_salary = np.std(df['Salary'])

# Print the calculated mean and standard deviation
print(f"Mean Salary: {mean_salary}")
print(f"Standard Deviation of Salary: {std_salary}")
```

```python
# Plot the distribution of salaries
plt.hist(df['Salary'], bins=10, edgecolor='black', alpha=0.7)

# Add a line for the mean salary
plt.axvline(mean_salary, color='red', linestyle='dashed', linewidth=2, label=f'Mean Salary: ${mean_salary:.2f}')

# Add lines for one standard deviation above and below the mean
plt.axvline(mean_salary - std_salary, color='blue', linestyle='dashed', linewidth=2, label=f'-1 Std Dev: ${mean_salary - std_salary:.2f}')
plt.axvline(mean_salary + std_salary, color='green', linestyle='dashed', linewidth=2, label=f'+1 Std Dev: ${mean_salary + std_salary:.2f}')

# Add titles and labels
plt.title('Salary Distribution')
plt.xlabel('Salary')
plt.ylabel('Frequency')
plt.legend()

# Show the plot
plt.show()
```
